# 1. Исследование на несмещенность

```python
# создание выборки из стандартного норм. распределения
def to_sample(n):
    n *= 2
    res = np.random.randn(n)
    return res
```

```python
# оценка дисперсии 
def to_estimate(sample):
    # сортировка выборки по возрастанию
    # создание двух массивов с четными и нечетными индексами
    # разница четных и нечетных поэлем-но, возведение каждого элемента в квадрат,
    # общая сумма, деленная на объем р-ия
    sorted_sample = sorted(sample)
    even = np.zeros(0)
    odd = np.zeros(0)
    
    for i in range(len(sorted_sample)):
        if i % 2 == 0:
            even = np.append(even, sorted_sample[i])
        else:
            odd = np.append(odd, sorted_sample[i])
    
    sm = np.sum((even - odd) ** 2)
    res = sm / (2 * n)
    return res
```

```python
# разные n для эксперимента
diff_n = [10, 50, 100, 200, 500, 1000]

for n in diff_n:
    experiments = 100   # повторение эксперимента 100 раз
    biases = np.zeros(0)
    
    for i in range(experiments):
        sample = to_sample(n)
        var_est = to_estimate(sample)
        true_var = 1   # истинная дисперсия для норм. р-ия = 1
        bias = var_est - true_var   # несмещенность для каждого n
        biases = np.append(biases, bias)
        
    print(f'Для {n} средняя несмещенность - {np.mean(biases)}')
```

# 2. Исследование на состоятельность

```python
# исследование состоятельности
def consistency(experiments, max_n):
    for n in range(1, max_n + 1):
        var_arr = np.zeros(0)
        
        for i in range(experiments):
            sample = to_sample(n)
            var_est = to_estimate(sample)
            var_arr = np.append(var_arr, var_est)
        # дисперсия оценки для каждого n
        disp = np.var(var_arr)
        print(f'Для n равному {n} дисперсия оценки равна {disp}')   
            
```

```python
# Для примера можно провести исследование для 1000 экспериментов, максимальное n = 100
consistency(experiments=1000, max_n=100)
```

### 1. С увеличением размера выборки дисперсия дисперсии оценки уменьшается => оценка становится все более точной при использовании все бОльших выборок.
### 2. Средняя несмещенность стремится к -1 при увеличении выборки => асимптотическая несмещенность оценки. Значит, оценка дисперсии является состоятельной и ассимпт. несмещенной.
